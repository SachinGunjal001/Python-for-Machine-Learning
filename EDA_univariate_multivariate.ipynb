# Scatterplot

```python
# ...existing code...
plt.figure(figsize=(15, 8))
sns.scatterplot(
    data=tips,
    x='total_bill',
    y='tip',
    hue='sex',            # column name, not the whole DataFrame
    style='smoker',
    size='size',          #
)
plt.title('Tip vs Total Bill')
plt.xlabel('Total bill')
plt.ylabel('Tip')
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', title='Legend')
plt.tight_layout()
plt.show()
```
