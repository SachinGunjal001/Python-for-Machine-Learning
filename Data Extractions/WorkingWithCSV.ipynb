```python
import requests
from io import StringIO
import pandas as pd

url = "https://raw.githubusercontent.com/SachinGunjal001/Market-Size-Analysis-on-Electric-Vehicles-Industries-/main/Electric_Vehicle_Population_Data.csv"

headers = {"User-Agent": "Mozilla/5.0"}

req = requests.get(url, headers=headers)
req.raise_for_status()  # will raise an error if something goes wrong

data = StringIO(req.text)
df = pd.read_csv(data)

print(df.head())
```
