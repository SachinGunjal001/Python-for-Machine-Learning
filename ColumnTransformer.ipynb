```python
import numpy as np
import pandas as pd

```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = pd.read_csv('covid_toy.csv')
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df.drop(columns=['has_covid']), df['has_covid'], test_size=0.2)
```

```python
x_train
```

# 1. Aam Zindagi

```python
si = SimpleImputer()
x_train_fever = si.fit_transform(x_train[['fever']])

x_test_fever = si.fit_transform(x_test[['fever']])
x_train_fever
```

```python
x_train_fever.shape
```

```python
#ordinal ecoding

oe = OrdinalEncoder(categories=[['Mild', 'Strong']])
x_train_cough = oe.fit_transform(x_train[['cough']])

x_test_cough = oe.fit_transform(x_test[['cough']])

x_train_cough.shape
```

```python
#onhot  ecoding

ohe = OneHotEncoder(drop='first', sparse_output=False)
x_train_gender_city = ohe.fit_transform(x_train[['gender','city']])

x_test_gender_city = ohe.fit_transform(x_test[['gender','city']])

x_train_gender_city.shape
```

```python
df.head()
```

```python
#extracting age
x_train_age = x_train.drop(columns=['gender', 'fever', 'cough', 'city'])

x_test_age = x_test.drop(columns=['gender', 'fever', 'cough', 'city'])
x_train_age.shape
```

```python
x_train_transformed = np.concatenate([x_train_age, x_train_fever, x_train_cough,x_train_gender_city], axis=1)

x_test_transformed = np.concatenate([x_test_age, x_test_fever, x_test_cough, x_test_gender_city], axis=1)

x_train_transformed
```

2. Mentos Zindagi

```python
from sklearn.compose import ColumnTransformer
```

```python
transformer = ColumnTransformer(transformers=[
    ('tnf1', SimpleImputer(), ['fever']),
    ('tnf2', OrdinalEncoder(categories=[['Mild', 'Strong']]), ['cough']),
    ('tnf3', OneHotEncoder(drop='first', sparse_output=False),['gender', 'city'])
], remainder='passthrough')
```

```python
transformer.fit_transform(x_train)
```
